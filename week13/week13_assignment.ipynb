### 2. Data Diversity

Use Python to identify how many unique values each column in your data has.  Use Python to identify which column has the greatest number of distinct values (aka the most diversity)?  If your data has a unique identifier (e.g. a person or visit or record ID) then find the column with the next most number of unique values.

```python
# Count the number of unique values in each column
unique_counts = heart_data.nunique()
```

```python
unique_counts
```

```python
# Find the column with the highest number of unique values
most_diverse_column = unique_counts.idxmax()
max_unique_values = unique_counts.max()

# Print the column with the most diverse values and the count of unique values
print(f"Column with the most diverse values: {most_diverse_column}")
print(f"Number of unique values in the column: {max_unique_values}")

# Drop the most diverse column to find the column with the second most unique values
unique_counts.drop(most_diverse_column, inplace=True)
second_max_unique_column = unique_counts.idxmax()
second_max_unique_count = unique_counts.max()

# Print the column with the second most diverse values and the count of unique values
print(f"\nColumn with the second most diverse values: {second_max_unique_column}")
print(f"Number of unique values in the column: {second_max_unique_count}")
```

### 3. Data Redundancy

Use Python to identify any columns in your data where the value for every row is the same.  For example, if you had a gender column and all the rows said 'M', then you would want to write the code that helped you determine that.

```python
# looping through all the columns to find if there is same value in entire column
for cols in heart_data:
    if (heart_data[cols] == heart_data[cols][0]).all():
        print("Columns with the same value for every row" +  cols, '->', heart_data[cols][0])
    else:
        print("No columns with the same value for every row in " + cols)
```

### 4. Range of Values

Your data almost certainly has some numeric columns that can be summed or averaged.  Create a histogram that shows the distribution of values for this column.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# List of columns for which distributions will be plotted
columns_of_interest = ["Age", "RestingBP"]

# Loop through the columns and create histograms
for column in columns_of_interest:
    plt.figure(figsize=(8, 6))  # Set the figure size
    sns.histplot(heart_data[column], kde=True, bins=15, color="skyblue", edgecolor="black")  # Create histogram with density curve
    plt.title(f"Distribution of {column}")  # Set the title for the plot
    plt.xlabel(column)  # Label for the X-axis
    plt.ylabel("Frequency")  # Label for the Y-axis
    plt.show()  # Display the plot
```

```python
# importing libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Heart Disease Distribution by Chest Pain Type
plt.figure(figsize=(10, 6))  # Set the figure size for the bar plot
sns.countplot(data=heart_data, x="ChestPainType", hue="HeartDisease", palette="Set2")  # Create count plot for chest pain type
plt.title("Heart Disease Distribution by Chest Pain Type")  # Title for the plot
plt.xlabel("Chest Pain Type")  # Label for the X-axis
plt.ylabel("Count")  # Label for the Y-axis
plt.show()

# Oldpeak vs Heart Disease
plt.figure(figsize=(10, 6))  # Set the figure size for the box plot
sns.boxplot(data=heart_data, x="HeartDisease", y="Oldpeak", hue="HeartDisease", palette="Set3")  # Create box plot
plt.title("Oldpeak Distribution by Heart Disease Status")  # Title for the plot
plt.xlabel("Heart Disease (0=No, 1=Yes)")  # Label for the X-axis
plt.ylabel("Oldpeak (ST Depression)")  # Label for the Y-axis
plt.show()

# Select only numeric columns
numeric_columns = heart_data.select_dtypes(include=["number"])
# Compute the correlation matrix
correlation_matrix = numeric_columns.corr()
# Plotting the heatmap
plt.figure(figsize=(10, 8))  # Set the size of the heatmap
sns.heatmap(
    correlation_matrix,
    annot=True,  # Annotate cells with correlation values
    cmap="coolwarm",  # Use a cool-warm color palette
    fmt=".2f",  # Format the correlation values to 2 decimal places
    linewidths=0.5,
)
plt.title("Correlation Heatmap")  # Title of the heatmap
plt.show()

# Cholesterol Distribution
plt.figure(figsize=(8, 6))  # Set the figure size for the histogram
sns.histplot(heart_data["Cholesterol"], kde=True, bins=20, color="orange", edgecolor="black")
plt.title("Cholesterol Distribution")  # Title for the plot
plt.xlabel("Cholesterol (mg/dL)")  # Label for the X-axis
plt.ylabel("Frequency")  # Label for the Y-axis
plt.show()

```
